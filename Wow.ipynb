```python
df['COSTING_PROJECTS']=df['COSTING_PROJECTS'].astype(str)
df['ECN_DESC']=df['ECN_DESC'].astype(str)
df['IMPACTED_ITEMS']=df['IMPACTED_ITEMS'].astype(str)

df_ecr['COSTING_PROJECTS']=df_ecr['COSTING_PROJECTS'].astype(str)
df_ecr['CHANGE DESCRIPTION']=df_ecr['CHANGE DESCRIPTION'].astype(str)
df_ecr['IMPACTEDITEM']=df_ecr['IMPACTEDITEM'].astype(str)
```

```python
import pandas as pd

def merge_ecr_to_df(df, df_ecr):
    """
    Merges rows of df_ecr to df based on matching criteria.
    """

    # Create temporary merge keys in both DataFrames
    df['merge_key'] = df['COSTING_PROJECTS'].astype(str) + df['ECN_DESC'].astype(str)
    df_ecr['merge_key'] = df_ecr['COSTING_PROJECTS'].astype(str) + df_ecr['CHANGE DESCRIPTION'].astype(str)

    # Merge the DataFrames on the merge key
    merged_df = pd.merge(df, df_ecr, on='merge_key', how='left', suffixes=('', '_ecr'))

    # Drop the temporary merge key column
    merged_df.drop(columns=['merge_key'], inplace=True)

    return merged_df

# Example usage
merged_df = merge_ecr_to_df(df, df_ecr)
```

```python
def find_missing_impacts(row):

  impacted_items = str(row['IMPACTED_ITEMS']).split(',')  # Split comma-separated values
  impacted_item = str(row['IMPACTEDITEM'])
  missing = [item for item in impacted_items if item != impacted_item and item != 'nan']
  # Find items not in 'IMPACTEDITEM' and not 'nan'

  return ','.join(missing)  # Join missing impacts with commas





merged_df['missing_impacts'] = merged_df.apply(find_missing_impacts, axis=1)
```
