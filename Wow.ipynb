```python
import pandas as pd

def merge_ecr_to_df(df, df_ecr):
    """
    Merges rows of df_ecr to df based on matching criteria.
    """

    # Create temporary merge keys in both DataFrames
    df['merge_key'] = df['COSTING_PROJECTS'].astype(str) + df['ECN_DESC'].astype(str)
    df_ecr['merge_key'] = df_ecr['COSTING_PROJECTS'].astype(str) + df_ecr['CHANGE DESCRIPTION'].astype(str)

    # Merge the DataFrames on the merge key
    merged_df = pd.merge(df, df_ecr, on='merge_key', how='left', suffixes=('', '_ecr'))

    # Drop the temporary merge key column
    merged_df.drop(columns=['merge_key'], inplace=True)

    return merged_df

# Example usage
merged_df = merge_ecr_to_df(df, df_ecr)
```

```python
def find_missing_impacts(row):

  impacted_items = str(row['IMPACTED_ITEMS']).split(',')  # Split comma-separated values
  impacted_item = str(row['IMPACTEDITEM'])
  missing = [item for item in impacted_items if item != impacted_item and item != 'nan']
  # Find items not in 'IMPACTEDITEM' and not 'nan'

  return ','.join(missing)  # Join missing impacts with commas





merged_df['missing_impacts'] = merged_df.apply(find_missing_impacts, axis=1)
```

```python
from __future__ import division
import ipywidgets as widgets
from IPython.display import display

import pandas as pd
import numpy as np
import re
import unicodedata

# Create input widgets
costing_projects_input = widgets.Text(description="JobNo:")
change_description_input = widgets.Text(description="Description:")
impacted_items_input = widgets.Text(description="Impacted Items:")
#division_input = widgets.Text(description="Division:")

search_button = widgets.Button(description="Search")

# Create output widget
output_widget = widgets.Output()



def normalize_text(text):
    text = unicodedata.normalize("NFKD", text)  # Normalize to remove special characters
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = text.strip()  # Remove leading/trailing spaces
    return text

merged_df["CHANGE DESCRIPTION"] = merged_df["CHANGE DESCRIPTION"].astype(str).apply(normalize_text)
merged_df["IMPACTED_ITEMS"] = merged_df["IMPACTED_ITEMS"].astype(str).apply(normalize_text)
merged_df["COSTING_PROJECTS"] = merged_df["COSTING_PROJECTS"].astype(str).apply(normalize_text)


# Define search function
def search_df(button):
  with output_widget:
    output_widget.clear_output()
    costing_projects = str(costing_projects_input.value)
    change_description = normalize_text(change_description_input.value)
    #division=division_input.value
    impacted_items = str(impacted_items_input.value)



    print(f"Input JobNo: {costing_projects}")  # Print normalized input JobNo
    print(f"Input Description: {change_description}")  # Print normalized input Description
    print(f"Input Impacted Items: {impacted_items}")  # Print normalized input Impacted Items




    # Filter the DataFrame
    pattern = '|'.join(impacted_items.split(','))  # Create regex pattern for any of the impacted items
    filtered_df = merged_df[
        (merged_df["COSTING_PROJECTS"] == costing_projects) &
        (merged_df["CHANGE DESCRIPTION"] == change_description) &
        #(merged_df["DIVISION"]==division)&
        (merged_df["IMPACTED_ITEMS"].str.contains(pattern, regex=True))
    ]



    if not filtered_df.empty:
          impacted_items_list = impacted_items.split(',')  # Split comma-separated values
          all_impacted_items = []
          for index in filtered_df.index:
            all_impacted_items.extend(str(filtered_df.loc[index, 'IMPACTED_ITEMS']).split(','))  # Get all impacted items from filtered rows

          # Find items not in impacted_items_list and not 'nan'
          remaining_missing = [item for item in all_impacted_items if item not in impacted_items_list and item != 'nan']
          remaining_missing = list(set(remaining_missing))  # Remove duplicates

          print(f"Remaining Missing Impacts: {','.join(remaining_missing)}")  # Join missing impacts with commas
    else:
          print("No matching records found.")
# search button

search_button.on_click(search_df)

# Display widgets
```
