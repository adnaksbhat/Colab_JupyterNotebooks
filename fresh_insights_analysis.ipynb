```python


import pandas as pd
import nltk
from collections import Counter
from itertools import chain
from itertools import combinations
import itertools
from nltk import ngrams

# Download required NLTK data
nltk.download('punkt')

def find_frequent_phrases(df, group_cols, text_col, min_phrase_length=2, top_n=10):
    frequent_phrases = {}
    for name, group in df.groupby(group_cols):
        all_text = ' '.join(group[text_col].astype(str).tolist())
        tokens = nltk.word_tokenize(all_text)
        phrases = chain.from_iterable(
            ngrams(tokens, n) for n in range(min_phrase_length, min(len(tokens) +1, 6)) # Limit ngram length
        )
        phrase_counts = Counter(phrases)

        frequent_phrases[tuple(name)] = phrase_counts.most_common(top_n)
    return frequent_phrases
```
