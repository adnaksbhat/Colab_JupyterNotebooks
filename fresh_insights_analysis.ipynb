```python
from collections import Counter
from itertools import chain

def find_frequent_phrases(df, group_cols, text_col, min_phrase_length=4, top_n=10):
    """
    Finds frequent phrases within groups based on other column values.

    Args:
        df: The pandas DataFrame.
        group_cols: A list of columns to group by.
        text_col: The column containing text data.
        min_phrase_length: The minimum length of phrases (number of words).
        top_n: The number of top frequent phrases to return.

    Returns:
        A dictionary mapping group values to frequent phrases and counts.
    """
    frequent_phrases = {}
    for name, group in df.groupby(group_cols):
        all_text = ' '.join(group[text_col].astype(str).tolist())
        words = all_text.split()
        phrases = chain.from_iterable(
            [' '.join(words[i:i + l]) for i in range(len(words) - l + 1)]
            for l in range(min_phrase_length, len(words) + 1)  # Vary phrase length from min_phrase_length to the maximum possible length
        )
        phrase_counts = Counter(phrases)
        frequent_phrases[tuple(name)] = phrase_counts.most_common(top_n)
    return frequent_phrases

# Example usage with minimum phrase length of 4:
group_cols = ['ACTIVITY_NAME', 'Train Composition', 'MATERIAL_CATEGORY']
frequent_phrases = find_frequent_phrases(df_eng2, group_cols, 'CHANGE_stmts', min_phrase_length=5, top_n=10)
```

```python
import re
def find_correlated_phrases(df, text_col, group_cols, min_phrase_length=4, top_n=10):
    """
    Finds frequent phrases in 'text_col' that correlate with consistent values in 'group_cols'.

    Args:
        df: The pandas DataFrame.
        text_col: The column containing text data (e.g., 'CHANGE DESCRIPTION').
        group_cols: A list of columns to check for consistent values (e.g., ['IMPACTEDITEM', 'MATERIAL_CATEGORY', etc.]).
        min_phrase_length: The minimum length of phrases (number of words).
        top_n: The number of top frequent phrases to return.

    Returns:
        A dictionary mapping phrases to correlated group values and counts.
    """
    correlated_phrases = {}
    for index, row in df.iterrows():
        description = row[text_col]
        words = description.split()
        phrases = chain.from_iterable(
            [' '.join(words[i:i + l]) for i in range(len(words) - l + 1)]
            for l in range(min_phrase_length, len(words) + 1)
        )
        for phrase in phrases:
            phrase = re.escape(phrase)
            # Find all rows with the same phrase
            matching_rows = df[df[text_col].str.contains(phrase, case=False, regex=True)]
            # Check if group_cols have consistent values in matching rows
            if matching_rows[group_cols].nunique(axis=0).sum() == len(group_cols):  # All columns have only one unique value
                group_values = tuple(matching_rows[group_cols].iloc[0].values)  # Get the consistent values
                correlated_phrases.setdefault(phrase, {}).setdefault(group_values, 0)
                correlated_phrases[phrase][group_values] += 1
    return correlated_phrases
```

```python
group_cols = ['IMPACTEDITEM', 'MATERIAL_CATEGORY', 'ACTIVITY_NAME', 'Train Composition']  # Columns to check for consistency
correlated_phrases = find_correlated_phrases(df_eng2, 'CHANGE_stmts', group_cols, min_phrase_length=4, top_n=10)
```

```python
for phrase, group_data in correlated_phrases.items():
    for group_values, count in group_data.items():
        if count > 5:  # Filter for patterns occurring more than once
            print(f"Phrase: '{phrase}'")
            print(f"Correlated with: {dict(zip(group_cols, group_values))}")
            print(f"Occurrences: {count}")
            print("-" * 20)
```
