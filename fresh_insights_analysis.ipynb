```python
import re
def find_correlated_phrases(df, text_col, group_cols, min_phrase_length=4, top_n=10):
    """
    Finds frequent phrases in 'text_col' that correlate with consistent values in 'group_cols'.

    Args:
        df: The pandas DataFrame.
        text_col: The column containing text data (e.g., 'CHANGE DESCRIPTION').
        group_cols: A list of columns to check for consistent values (e.g., ['IMPACTEDITEM', 'MATERIAL_CATEGORY', etc.]).
        min_phrase_length: The minimum length of phrases (number of words).
        top_n: The number of top frequent phrases to return.

    Returns:
        A dictionary mapping phrases to correlated group values and counts.
    """
    correlated_phrases = {}
    for index, row in df.iterrows():
        description = row[text_col]
        words = description.split()
        phrases = chain.from_iterable(
            [' '.join(words[i:i + l]) for i in range(len(words) - l + 1)]
            for l in range(min_phrase_length, len(words) + 1)
        )
        for phrase in phrases:
            phrase = re.escape(phrase)
            # Find all rows with the same phrase
            matching_rows = df[df[text_col].str.contains(phrase, case=False, regex=True)]
            # Check if group_cols have consistent values in matching rows
            if matching_rows[group_cols].nunique(axis=0).sum() == len(group_cols):  # All columns have only one unique value
                group_values = tuple(matching_rows[group_cols].iloc[0].values)  # Get the consistent values
                correlated_phrases.setdefault(phrase, {}).setdefault(group_values, 0)
                correlated_phrases[phrase][group_values] += 1
    return correlated_phrases
```

```python
# prompt: in df['material_category'] has values like "AUXILIARY BASEPLATE,CABLES,F2F"
# it has multiple values in a single cell. seperate each of these and add it into new row

import pandas as pd

def explode_column(df, column_name):
    """
    Explodes a column containing multiple values into separate rows.

    Args:
        df: The input pandas DataFrame.
        column_name: The name of the column to explode.

    Returns:
        A new pandas DataFrame with the exploded column.
    """
    df[column_name] = df[column_name].str.split(',')
    df = df.explode(column_name)
    df[column_name] = df[column_name].str.strip()
    return df

# Assuming 'df' is your DataFrame and 'material_category' is the column to explode
df_eng = explode_column(df_eng, 'MATERIAL_CATEGORY')
```

```python
# prompt: find correlation between CHANGE_stmts col and all other col
import pandas as pd
from dython.nominal import associations, cramers_v, theils_u

# Assuming 'df_eng' is your DataFrame

# Calculate Cramer's V for all pairs of columns including 'CHANGE_stmts'
def cramers_v_with_change_stmts(df, target_col='CHANGE_stmts'):
  """
  Calculates Cramer's V between a target column and all other columns in a DataFrame.

  Args:
    df: The pandas DataFrame.
    target_col: The name of the target column.

  Returns:
    A pandas Series containing Cramer's V values, indexed by column names.
  """
  cramers_v_values = {}
  for col in df.columns:
    if col != target_col:
      cramers_v_values[col] = cramers_v(df[target_col], df[col])
  return pd.Series(cramers_v_values).sort_values(ascending=False)



# Calculate and print Cramer's V
cramers_v_results = cramers_v_with_change_stmts(df_eng)
print("Cramer's V with CHANGE_stmts:")
print(cramers_v_results)
print("-" * 30)

```

```python
# prompt: # 2. Feature Engineering:
# #   a. Create new features based on the relationship:
# #      - Extract keywords from 'CHANGE_stmts' and create binary features indicating their presence
# #        for each 'IMPACTEDITEM' category.
# #      - Combine 'CHANGE_stmts' and 'IMPACTEDITEM' to form a new categorical feature.

import pandas as pd
import re

def add_keyword_features(df):
    """
    Extracts keywords from 'CHANGE_stmts' and creates binary features indicating their presence
    for each 'IMPACTEDITEM' category.  Combines 'CHANGE_stmts' and 'IMPACTEDITEM' into a new feature.

    Args:
        df: The input DataFrame.

    Returns:
        The DataFrame with new features.
    """

    # 1. Keyword Extraction and Binary Features
    keywords_by_impacted_item = {}
    for impacted_item in df['IMPACTEDITEM'].unique():
        subset = df[df['IMPACTEDITEM'] == impacted_item]
        keywords = []
        for change_stmt in subset['CHANGE_stmts']:
            if isinstance(change_stmt, str):  # Check if it is a string
              # Basic keyword extraction (can be improved)
                keywords.extend(re.findall(r'\b\w+\b', change_stmt.lower()))
        keywords_by_impacted_item[impacted_item] = set(keywords)  # Use sets for efficient checking

    for impacted_item, keywords in keywords_by_impacted_item.items():
        for keyword in keywords:
            df[f'keyword_{keyword}_{impacted_item}'] = df.apply(lambda row: 1 if keyword in str(row['CHANGE_stmts']).lower() else 0, axis=1)

    # 2. Combined Feature
    df['CHANGE_IMPACT'] = df['CHANGE_stmts'].astype(str) + '_' + df['IMPACTEDITEM'].astype(str)

    return df

# Example usage (assuming df_eng is your DataFrame)
df_eng = add_keyword_features(df_eng)
```
