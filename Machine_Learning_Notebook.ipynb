# K Means -> Clustering type |  Unsupervised Learning

```python
# unsupervised learning is used when we dont know class labels or target variable

# we take a k value

# clusters are formed around these centre placed k points

# sum of squared error (sse = summation(Xi-Ki)^2)    #Ki is centre point of cluster.   #Xi is all other points surrounding it

# upon plotting the graph for (K vs sse) u can see a elbow at certain point.  ----->.  """""this should be the value of K"""""
```

```python
import pandas as pd

df=pd.read_csv("/content/drive/MyDrive/Datasets/income.csv")

df.head()

#this dataset has no target col.  just random data to deal with.
#lets see if we can cluster it and find some interesting features
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.Age, df['Income($)'])
plt.xlabel("Age")
plt.ylabel("Income")
```

```python
from sklearn.cluster import KMeans

# how to consider k value??  solution -> elbow method

k_range=range(1,10)
sse=[]

for k in k_range:
  model=KMeans(n_clusters=k)
  model.fit(df[['Age', 'Income($)']])

  sse.append(model.inertia_)
```

```python
sse
```

```python
# now plot this sse

plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_range, sse)
```

```python
# therefore we get to know best k value is at elbow. That is 3
```

```python

model=KMeans(n_clusters=3)

model.fit(df[['Age', 'Income($)']])

y_predict = model.predict(df[['Age', 'Income($)']])

y_predict
```

```python
df['new_predict_cluster']=y_predict

df.head()
```

```python
df1=df[df.new_predict_cluster==0]
df2=df[df.new_predict_cluster==1]
df3=df[df.new_predict_cluster==2]



plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='black')

plt.x_label='Age'
plt.y_label='Income'
```

```python
# u can see some clusters have wrong points.   this is due to axis scaling issues.

# to solve this use min max scaling
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()


# we are scaling 'income' axis

scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])


# we are scaling 'age' axis
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])

df.head()
```

```python
# now do KMeans clustering again on scaled data


model=KMeans(n_clusters=3)

model.fit(df[['Age', 'Income($)']])
prediction_cluster=model.predict(df[['Age', 'Income($)']])

df['prediction_cluster']=prediction_cluster

df.head()
```

# Ensemble Learning | Bagging

```python
# to tackle high variance problem we use ensemble learning
# high variance means high difference between 2 models test error

# high variance happens when model overfits


#here we make predictions on different models and combine the result at the end

# bagging and boosting are 2 types
```

```python
#Bagging/Bootstrap

# samples of datset is splitted using replacement ensampling   ------------> this procedure is called bootstrap

#now on these sample datasets different model training is done.     ((((((( can use     SVM, KNN, Logistic.    )))))))

# all the results of diff models are averaged/majority vote decided ----------------> this procedure is aggregation
```

```python
# random forest is a bagging technique
# but here not only dataset rows are sampled, but also dataset cols
```

```python
import pandas as pd

df=pd.read_csv("/content/drive/MyDrive/Datasets/diabetes.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
X=df.drop('Outcome', axis='columns')
y=df['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_scaled=scaler.fit_transform(X)

X_scaled[:3]
#
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X_scaled, y, stratify=y, random_state=10)
```

```python
# lets see how decision tree works for this

from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import cross_val_score

scores=cross_val_score(DecisionTreeClassifier(), X, y, cv=5)

scores
```

```python
scores.mean()
```

```python
#okay. this model gives us 71% accuracy
```

```python
#now lets try using bagging technique

from sklearn.ensemble import BaggingClassifier

model= BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)


```

```python
# max_samples=0.8,  #use 80% of dataset
# oob_score=True       ->out of bag       -> there are some datapoints which wont appear in any samples. this datapoint is used for testing . get accuracy for that points. this is oob score
```

```python
model.fit(X_train, y_train)

model.score(X_test, y_test)

```

```python
model.oob_score_
```

```python
#using cross val

model= BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)

scores=cross_val_score(model, X, y, cv=5)

scores.mean()
```
