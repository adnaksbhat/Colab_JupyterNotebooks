#**Blackfriday sales dataset**

## preprocess cols. handle categorical cols

```python
import pandas as pd

df=pd.read_csv("/content/train.csv")

df.head()
```

```python
#drop userId and other irrelevent cols
df.drop(["User_ID","Product_ID"],axis=1,inplace=True)
df.head()
```

```python
#handle categorical cols
```

```python
#handle gender -> use labelencoding
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
df.head()
```

```python
#handle age -> use labelencoding
df["Age"].unique()
```

```python
df["Age"]=le.fit_transform(df["Age"])
df.head()
```

```python
#handle city_category
df["City_Category"]=le.fit_transform(df["City_Category"])
df.head()
```

```python
#handle Stay_In_Current_City_Years
df['Stay_In_Current_City_Years'].unique()
```

```python
#replace 4+ with 4
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].replace('4+','4')
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df.info()
```

```python
#Stay_In_Current_City_Years is of type object -> change to int
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

##data cleaning

```python
#handle missing values
df.isnull().sum()
```

```python
df['Product_Category_2'].value_counts()
```

```python
# we can see 8.0 as most freq value.
#so replace null with mode
```

```python
df['Product_Category_2'].mode()
```

```python
#so use 0th index
df['Product_Category_2'].mode()[0]
```

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df.isnull().sum()
```

##model building

```python
X=df.drop(["Purchase"],axis=1)
y=df["Purchase"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(n_estimators=10)
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```
